## Creating the dataframe without NLP or Neural Network components

```python
df2 = df1[['neighbourhood_group_cleansed','room_type','accommodates','bathrooms','bedrooms','beds',
          'bed_type','amenities','security_deposit','cleaning_fee','guests_included','extra_people',
          'minimum_nights', 'price']]
```

```python
df2.isnull().sum()
```

```python
#Replacing null values with 0 for security_deposit and cleaning_fee
df2['security_deposit'] = df2['security_deposit'].fillna(0)
df2['cleaning_fee'] = df2['cleaning_fee'].fillna(0)
```

```python
# cleaning dollar-amount columns
monetary_columns = ['security_deposit','cleaning_fee','extra_people','price']
for item in monetary_columns:
    df2[item] = df2[item].str.replace('$','').str.replace(',','').astype(float)
```

```python
# Dropping the few remaining units with null values
df2 = df2.dropna()
```

## Expanding "amenities" into boolean columns

```python
from collections import Counter
```

```python
# stripping the amenities column and creating a Counter of the most common amenities
results = Counter()
df['amenities'].str.strip('{}')\
               .str.replace('"', '')\
               .str.lstrip('\"')\
               .str.rstrip('\"')\
               .str.split(',')\
               .apply(results.update)

results.most_common(30)
```

```python
# making a list of amenities without count values
most_common = [i[0] for i in results.most_common(30)]
```

```python
# a for loop for creating boolean columns for each item in the most_common amenities
for item in most_common:
    df2[item] = df['amenities'].str.contains(item).astype(int)
```

```python
# Dropping 'amenities' because all information is contained in the engineered columns
df2 = df2.drop(columns='amenities')
```

```python
df2.head(1)
```

## Exploring price

```python
df2['price'].describe()
```

```python
# There are only 53/12,000 above 375. I think I can safely call those outliers
len(df2[df2['price'] > 375])
```

```python
# There are also a few set to 0, which doesn't make any sense
len(df2[df2['price'] == 0])
```

```python
# Removing all listings with a price of zero or greater than 375
df2 = df2[(df2['price'] > 0) & (df2['price'] < 375)]
```

## Separating into train-test split

```python
from sklearn.model_selection import train_test_split

train, val = train_test_split(df2, train_size=0.75, test_size=0.25, random_state=333)
```

```python
target = 'price'
target2 = ['security_deposit','cleaning_fee','extra_people']
X_train = train.drop(columns=[target])
y_train = train[target]
X_val = val.drop(columns=[target])
y_val = val[target]
```
